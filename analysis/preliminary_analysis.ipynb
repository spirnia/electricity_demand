```python
import pandas as pd
import numpy as np 
import matplotlib.pyplot as plt
df=pd.read_csv('C:\\workplace\\electricity_demand\\dataset\\AEP_hourly.csv')
```

```python
df.head()
```

```python
from statsmodels.graphics.tsaplots import plot_acf,plot_pacf
from statsmodels.tsa.seasonal import MSTL
```

```python

mstl = MSTL(df['AEP_MW'], periods=[24, 24 * 7,24*365])
res = mstl.fit()
```

```python
res.plot()
plt.figure(figsize=(40,40))
plt.show()
```

```python
res.seasonal.tail(24*3)['seasonal_24'].plot()
```

```python
res.seasonal.tail(24*28)['seasonal_168'].plot()
```

```python
res.trend.plot()
```

```python
plot_acf(df['AEP_MW'])
plt.show()
```
